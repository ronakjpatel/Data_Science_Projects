```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from sklearn.svm import SVC
```

```python
from sklearn.datasets import make_circles

#Creating the dataset
X, y = make_circles(n_samples=1000, random_state=123, noise=0.27, factor=0.2)

plt.figure(figsize=(8,6))

plt.scatter(X[y==0, 0], X[y==0, 1], color='red', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='green', alpha=0.5)
plt.title('Concentric circles')
plt.ylabel('y coordinate')
plt.xlabel('x coordinate')
plt.show()
```

```python
#creating the function for plotting

def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    
    
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
#function for executing Support Vector Machine algorithm with various parameters


def svm_algorithm (X_val,Y_val,kernal_name, C_val, random_state_val=0,gamma_val=None) :

  if gamma_val:

    # Create a SVC classifier using a linear kernel
    svm = SVC(kernel=kernal_name, gamma=gamma_val, C=C_val, random_state=random_state_val)
    # Train the classifier
    svm.fit(X_val,Y_val)
  else:
    # Create a SVC classifier using a linear kernel
    svm = SVC(kernel='linear', C=C_val, random_state=random_state_val)
    # Train the classifier
    svm.fit(X_val,Y_val)
  

  plt.scatter(X_val[Y_val==0, 0], X_val[Y_val==0, 1], color='red', alpha=0.5)
  plt.scatter(X_val[Y_val==1, 0], X_val[Y_val==1, 1], color='green', alpha=0.5)
  plt.title('Concentric circles')
  plt.ylabel('y coordinate')
  plt.xlabel('x coordinate')
  plt.figtext(0.5,1, "Kernal:"+kernal_name+", C:"+str(C_val)+", Gamma:"+str(gamma_val)+"(hyperparameters)", wrap=True, horizontalalignment='center', fontsize=20)
  plot_svc_decision_function(svm)









```

```python
svm_algorithm(X,y,'linear',1)
```
