```python
#importing libraries
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.model_selection
%matplotlib inline 
```

```python
#reading the dataset 
df = pd.read_csv('/Users/ronakpatel/Insurance_data.csv')
```

```python
#first split the data into Training and Testing Set

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2,random_state=25)
```

```python
#first we need to scale the  alues of the age bertween zero to one (divide by the 100)

X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
#cheking the values 
print(X_test)
print(X_test_scaled)
print(Y_test)
```

```python
#first build the simple NN model using keras

model = keras.Sequential([
    #kernel initilizer means the initial weighs in NN would be all ones
    keras.layers.Dense(1,input_shape=(2,),activation = 'sigmoid', kernel_initializer = 'ones', bias_initializer = 'zeros')
])

#after building the NN model now have to compile it 

model.compile(optimizer='adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])

#now it is the time for model fitting 

model.fit(X_train_scaled,Y_train,epochs=5000)
```

```python
model.evaluate(X_test_scaled,Y_test)
```

```python
def log_loss(y_true,y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
#sigmoid function for numpy array

def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age,affordability,y_true,epochs):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    
    
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        
        loss = log_loss(y_true,y_predicted)
        
        #we got the loss now we use derivative to find the the slope and can determine 
        # how and which way we can update the weights
        #derivative formulas for weight 1,2 and bias 
        w1d = (1/n) * np.dot(np.transpose(age),(y_predicted-y_true))
        w2d = (1/n) * np.dot(np.transpose(affordability),(y_predicted-y_true))
        bias_b = np.mean(y_predicted - y_true)
        
        #now it is time to update the wights 
        
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_b
        
        
        print (f'Epoch:{i}, weight_1:{w1}, weight_2:{w2}, bias:{bias}, loss:{loss}')
        
        
    return w1,w2,bias
        
    
            
        
```
